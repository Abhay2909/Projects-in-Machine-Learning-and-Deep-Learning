```python
import pandas as pd
import nltk
```

```python
train=pd.read_csv("train_movie_rating_prediction.csv")
test=pd.read_csv("test_movie_rating_prediction.csv")
```

```python
from nltk.tokenize import RegexpTokenizer
from nltk.stem.porter import PorterStemmer
from nltk.corpus import stopwords
```

```python
train.head()
```

```python
reviews=train['review']
```

```python
reviews=reviews.values
```

```python
print(reviews.shape)
```

```python
sw=set(stopwords.words('english'))
```

```python
ps=PorterStemmer()
```

```python
def cleanReview(review):
    
    sentence=review
    sentence=sentence.lower()
    
    sentence.replace("<br /><br />"," ")
    tokenizer=RegexpTokenizer('[a-z]+')
    tokens=tokenizer.tokenize(sentence)
    
    useful_tokens=[token for token in tokens if token not in sw]
    
    stemmed_tokens=[ps.stem(token) for token in useful_tokens]
    
    cleaned_review=' '.join(stemmed_tokens)
    
    return cleaned_review
```

```python
cleanReview(reviews[0])
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer()
```

```python
review_clean=[]
for i in range(reviews.shape[0]):
    review_clean.append(cleanReview(reviews[i]))
```

```python
vectorized_reviews=cv.fit_transform(review_clean)
```

```python
vectorized_reviews
```

```python
test.head()
```

```python
test_reviews=test['review']
```

```python
test_reviews=test_reviews.values
```

```python
print(test_reviews.shape)
```

```python
test_review_clean=[]
for i in range(test_reviews.shape[0]):
    test_review_clean.append(cleanReview(test_reviews[i]))
```

```python
test_vectorized_reviews=cv.transform(test_review_clean)
```

```python
from sklearn.naive_bayes import MultinomialNB,BernoulliNB, GaussianNB
```

```python
mnb=MultinomialNB()
print(mnb)
```

```python
from sklearn.preprocessing import LabelEncoder

le=LabelEncoder()
train["label"]=le.fit_transform(train["label"])
```

```python
y=train['label']
y=y.values
```

```python
y_train=[i for i in y]
print(len(y_train))
```

```python
mnb.fit(vectorized_reviews,y_train)
```

```python
y_test=mnb.predict(test_vectorized_reviews)
```

```python
print(len(y_test))
```

```python
y_test_actual=[]
for i in range(len(y_test)):
    if y_test[i]==0:
        y_test_actual.append('neg')
    else:
        y_test_actual.append('pos')
```

```python
y_test_actual
```

```python
final_list=[[index,value] for index,value in enumerate(y_test_actual)]
```

```python
final_list
```

```python
df1=pd.DataFrame(final_list)
```

```python
df1.to_csv('movie_rating_prediction_output.csv',index=False,header=['Id',"label"])
```
